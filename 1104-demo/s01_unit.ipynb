# Test Text Recognition

```python
import pytesseract
from PIL import Image
from pathlib import Path
```

```python
test_dir = Path().resolve() / 'data' / 'test'
test_img = test_dir / 'PICTs0054_new.JPG'

assert test_img.exists()

img = Image.open(test_img)
img
```

```python
text = pytesseract.image_to_string(img)
text
```

```python
out_dir = Path().resolve() / 'output'
# write to a file with the same name as the image file but with a .txt extension
out_file = out_dir / test_img.with_suffix('.txt').name
out_file.write_text(text)
```
